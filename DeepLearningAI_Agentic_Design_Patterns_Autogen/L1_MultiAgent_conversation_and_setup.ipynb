## Conversation

Setting up a conversation between two agents named Cathy and Joe where the memory of their conversations is retained.

```python
cathy = ConversableAgent(
    name="cathy",
    system_message=
    "Your name is Cathy and you are a Gen AI programmer who knows all aspects of compter engineering, machine learning and AI."
    "You are about to code a Support Vector Machine from scratch",
    llm_config=llm_config,
    human_input_mode="NEVER",
)

joe = ConversableAgent(
    name="joe",
    system_message=
    "Your name is Joe and you are a Gen AI programmer who knows all aspects of compter engineering, machine learning and AI."
    "Pick up the last piece of code and add to it",
    llm_config=llm_config,
    human_input_mode="NEVER",
)
```

```python
chat_result = joe.initiate_chat(
    recipient=cathy,
    message="I'm Joe. Cathy, how would you go about coding an SVM from scratch.",
    max_turns=10,
    verbose=True
)
```

### Better summarization of the conversation

```python
chat_result = joe.initiate_chat(
    cathy,
    message="I'm Joe. Cathy, let's keep on ideating about SVM",
    max_turns=2,
    summary_method="reflection_with_llm",
    summary_output="Summarize the conversation",
)
```

```python
pprint.pprint(chat_result.summary)
```

As we can see from above, one they have coded the SVM and dont have anything to add, the agents start talking non sense since they dont have anything left to talk about but are still under instructions to maintain the max_turns limit.
Hence it is necessary to allow them to end the conversation when they feel they dont have anything to add. This will come in with Chat Terminations

### Chat Termination

```python
cathy = ConversableAgent(
    name="cathy",
    system_message=
    "Your name is Cathy and you are a Gen AI programmer who knows all aspects of compter engineering, machine learning and AI."
    "You are about to code a Support Vector Machine from scratch"
    "When you are ready to end the conversation, say ' I gotta go.",
    llm_config=llm_config,
    human_input_mode="NEVER",
    is_termination_msg=lambda msg:"I gotta go" in msg["content"]
)

joe = ConversableAgent(
    name="joe",
    system_message=
    "Your name is Joe and you are a Gen AI programmer who knows all aspects of compter engineering, machine learning and AI."
    "Pick up the last piece of code and add to it"
    "When you are ready to end the conversation, say ' I gotta go.",
    llm_config=llm_config,
    human_input_mode="NEVER",
    is_termination_msg=lambda msg:"I gotta go" in msg["content"]
)
```

```python
chat_result = joe.initiate_chat(
    recipient=cathy,
    message="I'm Joe."
)
```

```python
chat_result = joe.initiate_chat(
    recipient=cathy,
    message="I'm Joe.I am willing to ideate SVMs with you, their basics and their coding aspects in python"
)
```

As we can see from above, both agents didnt end the conversation when they should have, meaning the boundary conditions of chat terminations need to be defined strictly. Lets try defining the boundary conditions of the chat termination strictly.


### Chat Termination with strict boundary conditions defined in prompt

```python
cathy = ConversableAgent(
    name="cathy",
    system_message=
    "Your name is Cathy and you are a Gen AI programmer who knows all aspects of compter engineering, machine learning and AI."
    "You are about to code a Support Vector Machine from scratch"
    "When you are ready to end the conversation, say ' I gotta go."
    # Strict boundary condition
    "Note that the moment you feel nothing new in technical terms is being added to the conversation, you need to end the conversation."
    "Also, if there is scope for something new to be added, dont end the conversation",    
    llm_config=llm_config,
    human_input_mode="NEVER",
    is_termination_msg=lambda msg:"I gotta go" in msg["content"]
)

joe = ConversableAgent(
    name="joe",
    system_message=
    "Your name is Joe and you are a Gen AI programmer who knows all aspects of compter engineering, machine learning and AI."
    "Pick up the last piece of code and add to it"
    "When you are ready to end the conversation, say ' I gotta go."
    # Strict boundary condition
    "Note that the moment you feel nothing new in technical terms is being added to the conversation, you need to end the conversation"
    "Also, if there is scope for something new to be added, dont end the conversation",    
    llm_config=llm_config,
    human_input_mode="NEVER",
    is_termination_msg=lambda msg:"I gotta go" in msg["content"]
)
```

```python
chat_result = joe.initiate_chat(
    recipient=cathy,
    message="I'm Joe.I am willing to ideate SVMs with you, their basics and their coding aspects in python"
)
```

As can be seen from above, the moment nothing new was being added, the agent identified it and left the conversation
